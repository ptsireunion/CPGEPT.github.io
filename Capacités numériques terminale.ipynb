# <font color=#FF6F7D > Taux d'avancement final d'une réaction</font>

### Aspects théoriques

<div class="alert alert-success" style="border-left: 15px solid #FF6F7D;border-radius:2px;">
    
**Rappel**  
Une réaction est considérée comme équilibrée lorsque l'avancement volumique final $x_f$ et l'avancement volumique maximal $x_\text{max}$ ne sont pas égaux :
$$x_f<x_\text{max}$$
Dans ce cas, on définit le taux d'avancement de la réaction comme le rapport de ces deux avancements :
$$\boxed{\tau=\dfrac{x_f}{x_\text{max}}}$$
Dans le cas d'une réaction quasi-totale, on considère ainsi que le taux d'avancement vaut 1.
</div>

Considérons la réaction entre l'acide éthanoïque et l'eau :
$$\text{CH}_3\text{COOH(aq)} + \text{H}_2\text{O}(\ell) = \text{CH}_3\text{COO}^-\text{(aq)} + \text{H}_3\text{O}^+\text{(aq)}$$

On considère la réaction d'une solution dont la concentration en acide éthanoïque est de $c_0$.

<summary>Le taux d'avancement de la réaction est donc donné par (<font color=#FF6F7D >cliquez ici pour avoir la réponse</font>) :

    $$\tau=\dfrac{x_f}{c_0}$$
    </details>


<details>
<summary>Le quotient de réaction de cette réaction est (<font color=#FF6F7D >cliquez ici pour avoir la réponse</font>) :  </summary>
    $$Q_r=\dfrac{[\text{CH}_3\text{COO}^-][\text{H}_3\text{O}^+]}{[\text{CH}_3\text{COOH}].c^\circ}$$
    </details>
On en déduit donc, à l'équilibre, d'après la loi d'action des masses :
$$K_A=\dfrac{[\text{CH}_3\text{COO}^-]_\text{éq}[\text{H}_3\text{O}^+]_\text{éq}}{[\text{CH}_3\text{COOH}]_\text{éq}.c^\circ}$$

Ainsi, en utilisant les grandeurs données, on obtient :
$$K_A=\dfrac{c_0.\tau^2}{(1-\tau).c^\circ}$$

On obtient ainsi l'équation du second degré suivante :
$$c_0.\tau^2+K_A.c^\circ.\tau+K_A.c^\circ=0$$
